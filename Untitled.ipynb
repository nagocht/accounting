```python
import os
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
```

```python
FILE_PATH = './data/Umsatzanzeige_DE70500105175419596903_20230917.csv'

def read_csv(file_path):
    df = pd.read_csv(file_path)
    return df
```

```python
df = pd.read_csv(FILE_PATH, sep=';',encoding='ANSI')
```

```python
df.dtypes
df_2023 = df[df['Buchung'].str.contains('2023')]
```

```python


def combine_columns(row):
    return '{}{}'.format(row['Auftraggeber/Empfaenger'], row['Verwendungszweck']).strip().lower().replace(" ", "")

def tag_transaction(row):
    string = row['combined']
    if any(substring in string for substring in rent):
        return 'rent'
    if any(substring in string for substring in phone):
        return 'phone'
    if any(substring in string for substring in health):
        return 'health'
    if any(substring in string for substring in groceries):
        return 'groceries'
    if any(substring in string for substring in dineout):
        return 'dineout'
    if any(substring in string for substring in transport):
        return 'transport'
    if any(substring in string for substring in vacation):
        return 'vacation'
    if any(substring in string for substring in salary):
        return 'salary'
    if any(substring in string for substring in cash):
        return 'cash'
    if any(substring in string for substring in sport):
        return 'sport'
    if any(substring in string for substring in household):
        return 'household'
    if any(substring in string for substring in clothes):
        return 'clothes'
    if any(substring in string for substring in freetime):
        return 'freetime'
    if any(substring in string for substring in internet):
        return 'internet'
    if any(substring in string for substring in insurance):
        return 'insurance'
    if any(substring in string for substring in electric):
        return 'electric'
    if any(substring in string for substring in wellness):
        return 'wellness'
    if any(substring in string for substring in finance):
        return 'finance'
    if any(substring in string for substring in education):
        return 'education'
```

```python
rent = ['wilhelmhilperthausverwaltunggmbh', 'wegspreekarree']
phone = ['hallon', 'drillisch']
education = ['coursra']
health = ['health', 'apotheke', 'seeger', 'Zahnaerztin', 'zahnarzt']
wellness = ['vabali', 'treatwell', 'dmdrogerie', 'dm-drogerie','ponyhuetchen', 'therme', 'kosmetik', 'hair', "rossmann"]
groceries = ['kaufland', 'rewe', 'penny', 'lidl', 'e-center', 'hellofresh', 'edeka', 'mpreis']
cash = ['bargeldauszahlung']
household = ['bauhaus', 'dimehouse', 'hoeffner']
dineout = ['baeckerei', 'coffee', 'cafe', 'cake', 'baguette', 'dits', 'brouwerij', 'southbeach', 'lwerk', 'haferkater', 'pita', 'thecommons', 'backer', 'takeaway',
           'crobag', 'subway', 'wieduwil', 'pizzawerk', 'rasthaus', 'raststaette', 'rastanlage', 'kneipphaeusl',
           'kitchen', 'gastronomie', 'sternenbaeck', 'backwerk', 'steinecke', 'pi-kant', 'standigevertretung', 'dbstore'
          ]
transport = ['bvg', 'dbvertrieb','dbfern', 'gvb', 'ns.nl', 'sixt', 'vasttrafik','emirates', 'logpay', 'rentalcars']
vacation = ['hotel', 'secretescapes', 'jointhecrew', 'resort']
salary = ['sharpist', 'lhkpotsdamfuerzbb']
sport = ['urbansports', 'decathlon']
clothes = ['kisura', 'momox', 'esprit', 'cloppenburg', 'otto', 'goertz', 'fashion', 'bestsecret']
freetime = ['cineplex', 'prime','tchibo', 'thalia', 'palastberlin', 'djh', 'sevenpeaks']
internet = ['vodafone']
insurance = ['versicherung', 'allianz', 'adac']
electric = ['vattenfall']
finance = ['ing']
subscription = ['bookbeat']
```

```python
df_2023['combined'] = df_2023.apply(lambda x: combine_columns(x), axis= 1)

# df
```

```python
df_2023['tag'] = df_2023.apply(lambda x: tag_transaction(x), axis= 1)
```
